```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
#Creating a model with 1 layer that has 1 neuron with 1 input shape
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
#Using stochastic gradient descent as optimizer and mean squared error as loss function
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#Adding some data
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
#Training the neural network
model.fit(xs, ys, epochs=500)
```

```python
#Using model to predict a new value of X
print(model.predict([10.0]))
```

The neural network didn't accurately exactly guess 19 as we would have assumed from the relationship y=2x-1 because I only fed in 6 data points. Neural networks deal with probabilities and therefore it was close to guessing y=2x-1 but doesn't know for sure.
